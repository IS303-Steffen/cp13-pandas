# Creating a Python Virtual Environment
Virtual environments fix a very common problem: different projects need different versions of the same packages. If you install everything globally (meaning on a single version of Python) on your computer, those versions can conflict and break things. A virtual environment is basically a private, isolated copy of Python for a single project. It has its own installed packages, separate from everything else on your system. That way, each project can independently have exactly what it needs without messing up your other projects.

You can make a virtual environment easily using VS Code. If you're interested, you can also make one from the terminal, using the tutorial below.

---
### Am I required to make a virtual environments for this class?
Technically, no. You can just install every package with pip globally. But learning about virtual environments will really help you in the long run if you ever have multiple projects you need to maintain across a few months or years.

# How to Create a Python Virtual Environment in VS Code

1. In VS Code, go to `View > Command Palette`
2. Type `Python: Create Environment`
3. Hit Enter/Return on your keyboard
4. Choose `Quick Create`
    - 
    - The Virtual Environment should now be created in the folder you currently have open in VS Code.
    - You should soon see a `.venv` folder in the file explorer in VS Code. That folder contains a new copy of python.
    - You could have also chosen the `venv` option instead of `Quick Create`, and that would let you name your folder something different, or use other versions of python. But for our purposes, `Quick Create` should always be good.
5. To make sure you are using your virtual environment when running Python code, click the trash can icon in your terminal, then go to `Terminal > New Terminal`.
    - Wait a couple seconds, and you should see `(.venv) in your terminal before your computer name. If you see that, you are good to go.

## BONUS INFO: Using the terminal to make a virtual environment

If you ever want to make a virutal environment but aren't using VS Code, you can just follow the steps below. The process is a little different for MacOS and Windows, so make sure you're following the right set of instructions:

---
### Step 0: Verify Python Is Installed

#### macOS
```bash
python3 --version
```

#### Windows
```powershell
py --version
```

---
## macOS Instructions

### 1. Navigate to Your Project Folder
```bash
cd path/to/your/project
```

### 2. Create the Virtual Environment
```bash
python3 -m venv .venv
```

### 3. Activate the Environment
```bash
source .venv/bin/activate
```

### 4. Install Packages
```bash
pip install requests
```

### 5. Deactivate
```bash
deactivate
```

---
## Windows Instructions

### 1. Navigate to Your Project Folder
```powershell
cd path\to\your\project
```

### 2. Create the Virtual Environment
```powershell
py -3 -m venv .venv
```

### 3. Activate the Environment

#### PowerShell
```powershell
.venv\Scripts\Activate.ps1
```

#### Command Prompt (cmd)
```cmd
.venv\Scripts\activate.bat
```

### 4. Install Packages
```powershell
pip install requests
```

### 5. Deactivate
```powershell
deactivate
```

---
## Important Notes

- Always activate the environment before installing packages.
- Use `python -m pip install ...` if you're unsure which interpreter pip is using.
- Add `.venv/` to your `.gitignore` file.
